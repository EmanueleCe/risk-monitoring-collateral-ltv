## Loan Structure and Collateral Allocation

The model assumes a single Lombard loan with a principal amount of **EUR 470 000**, granted to a high-net-worth private banking client. The loan is treated as a non-amortising exposure, with no scheduled repayments or additional drawdowns considered over the analysis horizon.

Rather than modelling a contractual maturity, the loan is assumed to be long-term in nature and continuously monitored from a risk management perspective. The focus of the analysis is therefore placed on the evolution of collateral value and loan-to-value (LTV) dynamics over time, rather than on repayment mechanics.

The pledged collateral portfolio reflects a growth-oriented but diversified asset allocation, combining liquidity, capital preservation, and exposure to equity markets. The initial collateral allocation is defined as follows:

- **10% Cash / Money Market instruments**, providing immediate liquidity and stability  
- **40% Investment Grade Bond ETF**, contributing income and downside protection  
- **50% Broad Market Equity ETF**, representing the primary source of growth and return  

Collateral weights are assumed to remain constant throughout the analysis. No rebalancing, substitution of assets, or changes in portfolio composition are allowed, ensuring that observed changes in risk metrics are driven exclusively by market price movements.

This structure results in an initial loan-to-value ratio comfortably below the margin call threshold, while still allowing for meaningful stress scenarios and risk escalation under adverse market conditions.



## Loan-to-Value (LTV) Monitoring, Margin Calls and Liquidation

In Lombard lending, risk monitoring is centered on the loan-to-value (LTV) ratio, which measures how well the outstanding loan exposure is covered by the pledged collateral. The LTV is defined as the ratio between the loan amount and the haircut-adjusted value of the collateral portfolio.

Unlike products such as residential mortgages, where affordability and borrower income play a key role, Lombard loans are primarily collateral-driven. The main source of risk is therefore not the borrowerâ€™s ability to service the loan, but the potential decline in the market value of the collateral. For this reason, the evolution of LTV over time represents the core risk indicator for the lender.

### Daily LTV Monitoring

Collateral values and LTV ratios are monitored on a daily basis using end-of-day market prices. Each day, the haircut-adjusted value of the collateral is recalculated and compared against the outstanding loan exposure.

Risk escalation is not driven by the size of daily market movements, but by the absolute level of LTV reached. Even small changes in market prices can trigger risk actions if predefined LTV thresholds are breached. As a result, there is no minimum percentage change or daily variation requirement for a risk event to occur.

### Margin Call Mechanism

A margin call is triggered when the LTV exceeds a predefined margin call threshold. In this model, the margin call threshold is set at **75% LTV**.

When this level is breached, the loan is still considered recoverable, but corrective action is required. The borrower is expected to restore adequate collateral coverage, typically by posting additional collateral or partially repaying the loan. The margin call therefore acts as an early warning mechanism, allowing risk to be addressed before the situation deteriorates further.

### Liquidation Mechanism

A second and higher LTV threshold defines the liquidation level, which represents a hard risk limit for the lender. In this model, the liquidation threshold is set at **90% LTV**.

If the LTV exceeds this level, the lender is no longer reliant on borrower action. Instead, it may initiate forced liquidation of part of the pledged collateral in order to reduce the loan exposure and restore sufficient collateral coverage. Liquidation is intended as a protective measure of last resort, aimed at preventing losses in the event of severe market deterioration.

### Risk Escalation Logic

The presence of two separate thresholds reflects standard risk management practice in collateralised lending. Margin calls provide an opportunity for preventive intervention, while liquidation represents a mandatory protective action. Both mechanisms are triggered exclusively by the level of LTV observed and are independent of the speed or magnitude of daily market movements.
