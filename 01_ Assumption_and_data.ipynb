**Lombard loans** are secured credit facilities typically offered to private wealth management clients, where the loan exposure is backed by a portfolio of liquid, marketable financial assets.

Unlike traditional unsecured lending, the primary source of risk in Lombard lending is not borrower creditworthiness, but fluctuations in the market value of the pledged collateral.

To mitigate this risk, collateral values are required to exceed the outstanding loan amount at all times. Risk control is achieved through the application of conservative haircuts, continuous monitoring of loan-to-value (LTV) ratios, and predefined margin call and liquidation mechanisms.

The appropriate LTV level depends on the type, liquidity, and volatility of the underlying collateral. As market prices evolve, changes in collateral value directly affect LTV, potentially triggering margin calls or forced liquidation if predefined thresholds are breached.

Haircuts represent an additional buffer applied to collateral values in order to account for market volatility, liquidity risk, and potential delays in liquidation. Assets with higher volatility or lower liquidity are subject to more conservative haircuts.

In recent years, digital assets such as Bitcoin have increasingly been used as Lombard collateral. While offering high liquidity, their elevated volatility typically results in lower admissible LTV levels and stricter risk controls compared to traditional financial instruments.

## Model Scope and Objective

The objective of this project is to develop a simplified yet realistic risk monitoring framework for Lombard loans, focusing on the dynamics between loan exposure and collateral value over time.

The model is designed to simulate how a financial institution monitors collateralised lending positions through loan-to-value (LTV) metrics, margin call triggers, and liquidation thresholds under normal and stressed market conditions.

The scope of the analysis is intentionally limited to collateral-driven risk. Borrower creditworthiness, loan pricing, regulatory capital, and provisioning frameworks are explicitly excluded in order to maintain a clear focus on market risk and collateral management mechanics.

The model operates under a set of transparent and static assumptions, allowing for clear interpretation of results and ensuring that changes in risk metrics are directly attributable to market movements in the underlying collateral.

This project is intended as a portfolio-grade analytical exercise, reflecting the structure and logic of internal risk monitoring tools used in private banking and wealth management environments.

## Key Model Assumptions

The model assumes a single, static Lombard loan with a fixed principal amount denominated in EUR. The loan is non-amortising over the analysis horizon, and no additional drawdowns or repayments are considered.

The collateral pool consists of a diversified portfolio of liquid, marketable financial assets, such as equity and bond exchange-traded funds (ETFs). Collateral composition and asset weights are assumed to remain constant over time.

Conservative, asset-specific haircuts are applied to collateral values in order to account for market volatility and liquidity risk. Haircut levels are fixed throughout the simulation and reflect internal risk management policies rather than regulatory requirements.

Collateral and loan risk are monitored on a daily basis using end-of-day market prices. Loan-to-value (LTV) ratios are recalculated daily, and predefined margin call and liquidation thresholds are applied mechanically whenever breaches occur.

Interest accrual on the loan, funding costs, and borrower servicing behavior are outside the scope of this model and are assumed to be managed independently from the collateral monitoring process.

The model focuses exclusively on collateral-driven risk dynamics. Borrower credit quality, pricing considerations, regulatory capital, and accounting frameworks are intentionally excluded.

## Collateral Portfolio and Risk Policy

The collateral portfolio considered in this model is intended to resemble a typical diversified allocation of a high-net-worth (HNW) private banking client. Such portfolios are generally structured to combine liquidity, capital preservation, and long-term growth, while remaining sufficiently robust to support secured lending activities.

From a risk management perspective, the primary objective of the collateral structure is not return maximisation, but rather the preservation of value under adverse market conditions. For this reason, the collateral mix includes assets with different risk profiles and liquidity characteristics, allowing the lender to mitigate concentration risk and reduce sensitivity to market shocks affecting a single asset class.

### Haircut Policy

Haircuts are applied to the market value of each collateral asset in order to provide an additional safety buffer against market volatility, liquidity constraints, and potential delays in collateral liquidation. Assets characterised by higher price volatility or lower liquidity are subject to more conservative haircuts, while more stable and liquid instruments receive lower adjustments.

Haircut levels are assumed to be conservative and remain constant for the duration of the simulation. They are intended to reflect internal risk management practices typically applied in private banking environments, rather than regulatory capital or accounting requirements. This choice allows the model to focus on economic risk dynamics rather than compliance-driven considerations.

### Loan-to-Value (LTV) Risk Thresholds

Collateral risk is monitored through the loan-to-value (LTV) ratio, defined as the ratio between the outstanding loan amount and the haircut-adjusted value of the pledged collateral. The LTV represents the central risk indicator used to assess the adequacy of collateral coverage over time.

Two distinct LTV thresholds are introduced to reflect different stages of risk escalation. The first threshold corresponds to a margin call level, at which the borrower is expected to take corrective action by either posting additional collateral or reducing the outstanding loan exposure. This mechanism serves as an early warning signal, allowing risk to be addressed before it becomes critical.

A second, higher threshold represents the liquidation level. If breached, the lender is no longer reliant on borrower action and may initiate forced liquidation of part of the collateral portfolio in order to restore the loan to acceptable risk levels. The use of separate margin call and liquidation thresholds reflects standard risk management practice, where preventive measures precede mandatory intervention.


### Monitoring Frequency

Collateral values and LTV ratios are monitored on a daily basis using end-of-day market prices. This frequency reflects standard operational practice in collateralised lending, where daily valuation allows for timely identification of risk deterioration and prompt escalation when predefined thresholds are breached.

### Scope Clarifications

For clarity and focus, several elements are intentionally excluded from the scope of this model. Interest accrual on the loan, borrower servicing behaviour, and funding costs are assumed to be managed independently and are not explicitly modelled. Similarly, borrower credit quality, loan pricing, regulatory capital, and accounting frameworks fall outside the scope of the analysis.

The model is designed to focus exclusively on collateral-driven risk dynamics and monitoring logic, providing a transparent and interpretable framework for understanding how market movements affect the risk profile of a Lombard loan.
